```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as plticker
```

```python
# resample = pd.read_csv('cifar10-5shard-CNN-fall-trimmed-mean-resample_output.csv')
# random = pd.read_csv('cifar10-5shard-CNN-fall-trimmed-mean-random_output.csv')
no_defense = pd.read_csv('shakespeare-fall-no-defense_output.csv')
# zeno = pd.read_csv('cifar10-5shard-CNN-fall-trimmed-mean_output.csv')
g3 = pd.read_csv('shakespeare-fall-zeno-outer-resample-3g_output.csv')
g5 = pd.read_csv('shakespeare-fall-zeno-outer-resample-5g_output.csv')
g10 = pd.read_csv('shakespeare-fall-zeno-outer-resample-10g_output.csv')
no_attack = pd.read_csv('shakespeare_output.csv')
# zeno.head()
```

```python
n = 10
print(g3[-n:].test_acc.std(), g5[-n:].test_acc.std(), g10[-n:].test_acc.std(), no_attack[-n:].test_acc.std(), no_defense[-n:].test_acc.std())
print(g3[-n:].test_acc.mean(), g5[-n:].test_acc.mean(), g10[-n:].test_acc.mean(), no_attack[-n:].test_acc.mean(), no_defense[-n:].test_acc.mean())
```

```python
plt.figure(figsize=(6,4), dpi=120)
# plt.gca().yaxis.set_major_locator(plticker.MultipleLocator(base=0.05))
plt.plot(g3.index, g3.test_acc, label='group size 3')
plt.plot(g5.index, g5.test_acc, label='group size 5')
plt.plot(g10.index, g10.test_acc, label='group size 10')
plt.plot(no_defense.index, no_defense.test_acc, label='no defense')
plt.plot(no_attack.index, no_attack.test_acc, label='no attack')
plt.title('CIFAR-10 IID with -g and Zeno')
plt.legend()
plt.show()
```
