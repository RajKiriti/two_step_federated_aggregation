```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as plticker
```

```python
resample = pd.read_csv('cifar10-iid-CNN-10fall-zeno-outer-resample_output.csv')
random = pd.read_csv('cifar10-iid-CNN-10fall-zeno-random_output.csv')
no_defense = pd.read_csv('cifar10-iid-CNN-10fall-no-defense_output.csv')
zeno = pd.read_csv('cifar10-iid-CNN-10fall-zeno_output.csv')
no_attack = pd.read_csv('cifar10-iid-CNN_output.csv')
# zeno.head()
```

```python
n = 10
print(resample[-n:].test_acc.std(), random[-n:].test_acc.std(), zeno[-n:].test_acc.std(), no_defense[-n:].test_acc.std())
print(resample[-n:].test_acc.mean(), random[-n:].test_acc.mean(), zeno[-n:].test_acc.mean(), no_defense[-n:].test_acc.mean())
```

```python
plt.figure(figsize=(8,6), dpi=120)
plt.gca().yaxis.set_major_locator(plticker.MultipleLocator(base=0.05))
plt.plot(resample.index, resample.test_acc, label='resample')
plt.plot(random.index, random.test_acc, label='random')
plt.plot(zeno.index, zeno.test_acc, label='zeno')
plt.plot(no_defense.index, no_defense.test_acc, label='no defense')
plt.plot(no_attack.index, no_attack.test_acc, label='no attack')
plt.title('IID with -10g')
plt.legend()
plt.show()
```
